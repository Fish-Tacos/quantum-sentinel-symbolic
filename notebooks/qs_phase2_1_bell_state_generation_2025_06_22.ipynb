```python
from qiskit import Aer, execute, QuantumCircuit
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt

# Define basis map
basis_map = {0: 'Z', 1: 'X', 2: 'Y'}

# Generate first N Fibonacci numbers
def fibonacci(n):
    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

# Map Fibonacci sequence to basis symbols (mod 3)
def symbolic_bases_from_fibonacci(n):
    fib_seq = fibonacci(n)
    return [basis_map[i % 3] for i in fib_seq]

# Create Bell circuit with flexible bases
def create_bell_with_basis(q0_basis, q1_basis):
    qc = QuantumCircuit(2, 2)
    qc.h(0)
    qc.cx(0, 1)

    for qubit, basis in zip([0, 1], [q0_basis, q1_basis]):
        if basis == 'X':
            qc.h(qubit)
        elif basis == 'Y':
            qc.sdg(qubit)
            qc.h(qubit)
        # Z needs no transformation

    qc.measure([0, 1], [0, 1])
    return qc

# Run multiple experiments using Fibonacci basis pattern
num_experiments = 10
q0_bases = symbolic_bases_from_fibonacci(num_experiments)
q1_bases = symbolic_bases_from_fibonacci(num_experiments)[::-1]  # reverse for variety

# Accumulate counts
total_counts = {}

simulator = Aer.get_backend('qasm_simulator')
for i in range(num_experiments):
    qc = create_bell_with_basis(q0_bases[i], q1_bases[i])
    job = execute(qc, simulator, shots=512)
    result = job.result()
    counts = result.get_counts()

    for outcome, count in counts.items():
        total_counts[outcome] = total_counts.get(outcome, 0) + count

# Show result
print("Total counts across Fibonacci-scheduled experiments:", total_counts)
plot_histogram(total_counts)
```

```python
import csv, math
import os
from datetime import datetime
from qiskit import QuantumCircuit, Aer, execute

# Maps symbolic numbers to measurement bases
basis_map = {0: 'Z', 1: 'X', 2: 'Y'}

# --- Symbol Generators ---

def fibonacci(n):
    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

def prime_numbers(n):
    primes = []
    num = 2
    while len(primes) < n:
        for p in primes:
            if num % p == 0:
                break
        else:
            primes.append(num)
        num += 1
    return primes

# --- Base Mapping from Symbolic Sequence ---
def symbolic_bases_from(symbol_type, n):
    if symbol_type == "fibonacci":
        seq = fibonacci(n)
    elif symbol_type == "primes":
        seq = prime_numbers(n)
    else:
        raise ValueError("Unsupported symbolic type.")

    return [basis_map[i % 3] for i in seq]

# --- Entropy Calculation ---
def shannon_entropy(counts):
    total = sum(counts.values())
    probs = [v / total for v in counts.values()]
    return -sum(p * math.log2(p) for p in probs if p > 0)

# --- Bell State Preparation and Measurement ---
def create_bell_with_basis(q0_basis, q1_basis):
    qc = QuantumCircuit(2, 2)
    qc.h(0)
    qc.cx(0, 1)
    for qubit, basis in zip([0, 1], [q0_basis, q1_basis]):
        if basis == 'X':
            qc.h(qubit)
        elif basis == 'Y':
            qc.sdg(qubit)
            qc.h(qubit)
    qc.measure([0, 1], [0, 1])
    return qc

# --- Run and Log Batch ---
def run_symbolic_batch(symbol_type="fibonacci", num_experiments=10, observer_note=""):
    log_file = f"quantum_log_{symbol_type}.csv"
    with open(log_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['timestamp', 'run_id', 'symbol_type', 'q0_basis', 'q1_basis', 'counts', 'entropy', 'observer_note'])

    q0_bases = symbolic_bases_from(symbol_type, num_experiments)
    q1_bases = symbolic_bases_from(symbol_type, num_experiments)[::-1]

    simulator = Aer.get_backend('qasm_simulator')

    for i in range(num_experiments):
        qc = create_bell_with_basis(q0_bases[i], q1_bases[i])
        job = execute(qc, simulator, shots=512)
        result = job.result()
        counts = result.get_counts()
        entropy = shannon_entropy(counts)

        print(f"Logging run {i+1}")
        print(f"q0_basis: {q0_bases[i]}, q1_basis: {q1_bases[i]}")
        print(f"counts: {counts}")
        print(f"entropy: {entropy}")

    with open(log_file, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([
            datetime.utcnow().isoformat(),
            f"{symbol_type}_{i+1:04}",
            symbol_type,
            q0_bases[i],
            q1_bases[i],
            str(counts),
            round(entropy, 5),
            observer_note
        ])
        f.flush()
        os.fsync(f.fileno())  # <--- ✅ This forces the data to be saved immediately


```
